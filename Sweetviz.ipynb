# Importing the Dependencies

```python
# Importing required libraries

import pandas as pd
import sweetviz
```

```python
# Reading the training dataset

train = pd.read_csv("/content/train (house).csv")
train.head()
# Reading the testing dataset

test = pd.read_csv("/content/test (house).csv")
test.head()
```

# Analyzing the dataset using Sweetviz Library

```python
my_report = sweetviz.analyze([train, "Train"], target_feat="SalePrice")
```

# Let's see the Whole report

```python
### create a whole report in form of HTML file

my_report.show_html("Report.html")
```

# Lets compare the training and testing dataset using Sweetviz

```python
# In case of testing data pass separate list for testing data as parameter

my_report1 = sweetviz.compare([train, "Train"], [test, "Test"], "SalePrice")
my_report1.show_html("Report2.html")
```
